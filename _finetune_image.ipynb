```python
from diffusers import StableDiffusionPipeline
import torch
from IPython.display import Image, display
from PIL import Image as PILImage
from PIL import ImageFilter

# Load the updated model
pipe = StableDiffusionPipeline.from_pretrained("/content/output_gttc_plus_prod", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Test prompts
prompts = [
    "a photo of [GTTC] building at night",
    "a photo of [PROD] product on a shelf"
]

for prompt in prompts:
    image = pipe(prompt, num_inference_steps=100, guidance_scale=10).images[0]
    pil_image = image.filter(ImageFilter.SHARPEN)  # Sharpen
    pil_image = pil_image.resize((256, 256), PILImage.LANCZOS)  # Resize
    filename = f"gttc_{prompt.split()[3]}.png"
    pil_image.save(filename)
    print(f"Image saved as '{filename}'")
    display(Image(filename=filename))
```
